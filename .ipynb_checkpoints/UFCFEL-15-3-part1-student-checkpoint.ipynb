```python
### Load in the libraries and the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx

def load_csv_data():
    data = pd.read_csv('./part1/packet-capture3.csv')
    return data

data = load_csv_data()
data = data[ data['Source'].isin(data['Source'].value_counts().head(30).index) ] 
data = data.drop('No.', axis=1)
data
```

### Question 1: Plot a Line Chart that clearly depicts "Seconds" on the x-axis, and "Total Number of Packets" sent on the y-axis, complete with axis labeling. 

#### (4 marks)

*Hint: The time is given in seconds with millisecond precision - think how you could aggregate this value to represent seconds only, then research a function that counts the value totals in your data.*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Convert time to seconds (rounding down)
data['Time'] = np.floor(data['Time']).astype(int)

# Count the number of packets per second
packets_per_second = data.groupby('Time').size()

# Plot
plt.figure(figsize=(10, 6))
packets_per_second.plot(kind='line')
plt.title('Total Number of Packets Sent Per Second') # Chart Title
plt.xlabel('Seconds') # x axis
plt.ylabel('Number of Packets') # y axis
plt.grid(True)
plt.show()
```

### Question 2: Display a Bar Chart that shows "Protocol" on the x-axis, and "Count" on the y-axis, complete with axis labeling. 
#### (2 marks)

*Hint: Search the pandas documentation for creating a Bar Chart from a DataFrame column.*

```python
# ANSWER
import matplotlib.pyplot as plt

protocol_counts = data['Protocol'].value_counts()


protocol_counts.plot(kind='bar') # plotting a bar chart 

# Adding labels to the axes
plt.xlabel('Protocol') # x axis
plt.ylabel('Count') # y axis

# Title of chart
plt.title('Count of Each Protocol') 
 
# Display the plot
plt.grid(axis='y')  # Adding grid lines for better readability
plt.show()

plt.show()
```

### Question 3: Display a Scatter Chart that shows the association between Source and Destination data. 

#### (2 marks)
*Hint: Matplotlib has a scatterplot function that takes x and y as inputs*

```python
# ANSWER
plt.figure(figsize=(20,10))
plt.xticks(rotation=20)
plt.yticks(rotation=20)
import matplotlib.pyplot as plt

# grabbing destination from source 
plt.scatter(data['Source'], data['Destination'])

# Adding labels to the axes
plt.xlabel('Source') # x axis 
plt.ylabel('Destination') # y axis

# Adding the  title to the plot
plt.title('Scatter Chart: Source vs Destination')

# Display the plot
plt.grid(True)
plt.show()

plt.show()
```

### Question 4: Display the data as a node link diagram using Source and Destination as your edges. 

#### (3 marks)

*Hint: Look at the NetworkX library: https://networkx.org/ and the online course notes.*

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Creating the graph itself
G = nx.Graph()


for _, row in data.iterrows():
    G.add_edge(row['Source'], row['Destination'])

# Draw the graph with green node coloyr 
plt.figure(figsize=(12, 8))
nx.draw(G, with_labels=True, node_size=700, node_color='lightgreen', font_size=10, font_weight='bold')

# Adding the title 
plt.title('Node-Link Diagram of Source and Destination')

# Display the plot
plt.show()
```

### Question 5: For each Protocol type, create a new Column and assign whether the Protocol usage is True or False 

#### (2 marks)
Hint: Get a list of unique protocol values, assign each value to be a new column where the Protocol column is equal to the Protocol name.

```python

import pandas as pd
from tabulate import tabulate

unique_protocols = data['Protocol'].unique()

# Creating new columns for each protocol
for protocol in unique_protocols:
    data[protocol] = data['Protocol'] == protocol

# Displaying the different columns 
display_columns = ['Time', 'Source', 'Destination', 'Protocol'] + list(unique_protocols)[:5]
tidy_data = data[display_columns]

# Displaying the first 40 rows of the updated DataFrame in a compact table format
print(tabulate(tidy_data.head(40), headers='keys', tablefmt='pipe', numalign="center"))


```

### Question 6: Show a Multi-Line Chart that shows the Total Number of Packets Per Protocol on the y-axis and "Seconds" on the x-axis. You should display a legend that states the colour of each protocol.

#### (6 marks)
*Hint: Think about how you did this in Question 1 , and recall that plt.plot can be used to append to an existing plot.*

```python

# ANSWER
import pandas as pd
import matplotlib.pyplot as plt

# Convert time to seconds 
data['Time'] = data['Time'].astype(int)

# Grouping the data by 'Time' 'Protocol' and count packets
grouped_data = data.groupby(['Time', 'Protocol']).size().reset_index(name='Count')

# Plotting the chart
plt.figure(figsize=(12, 6))

# Iterate through each protocol and plotting
for protocol in data['Protocol'].unique():
    protocol_data = grouped_data[grouped_data['Protocol'] == protocol]
    plt.plot(protocol_data['Time'], protocol_data['Count'], label=protocol)

plt.title('Total Number of Packets Per Protocol Over Time') # title
plt.xlabel('Seconds') # x axis 
plt.ylabel('Total Number of Packets') # y axis
plt.legend(title='Protocol')
plt.grid(True)
plt.show()
```
